```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv')
df
```

```python
#1
edu = df.groupby('Educational Attainment')['Population Count'].sum()
edu
```

```python
#2
percentage = df.loc[df['Age'] != '00 to 17'].groupby('Educational Attainment').sum()
percentage['PERCENT DISTRIBUTION'] = percentage['Population Count'] / \
percentage['Population Count'].sum() * 100
percentage
```

```python
#3
temp=df.groupby(['Educational Attainment','Personal Income'],as_index=False).sum("Population Count")
pivot=temp.pivot(index='Personal Income',columns='Educational Attainment',values="Population Count")
pivot
```

```python
#4
pivot.plot(kind="bar")
```
