NULL HYPOTHESIS : Prices of poroducts with ratings 4.3 to 4.5 is same as the price of products with ratings 4.5 to 5.

ALTERNATE HYPOTHESIS : Prices of poroducts with ratings 4.3 to 4.5 is NOT same as the price of products with ratings 4.5 to 5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/amazon.csv')
df1 = pd.read_csv('/content/amazon.csv')
```

```python
df['Stars'] = df['Stars'].str.strip('out of 5 stars')
```

```python
df['Stars'] = pd.to_numeric(df['Stars'],errors='coerce')
```

```python
df['Discounted_Price'] = df['Discounted_Price'].str.strip('₹')
```

```python
df['Discounted_Price'] = df['Discounted_Price'].str.strip(',')
```

```python
df['Discounted_Price'] = pd.to_numeric(df['Discounted_Price'],errors='coerce')
```

```python
stars1 = df.loc[(df['Stars'] >= 4.5)
                     & (df['Stars'] <= 5)]
stars1
```

```python
stars2 = df.loc[(df['Stars'] >= 4.3)
                     & (df['Stars'] <= 4.4)]
stars2
```

```python
observed_difference = stars2['Discounted_Price'].mean() - stars1['Discounted_Price'].mean()
observed_difference
```

```python
fig = plt.figure(figsize=(18, 18))
stars1['Discounted_Price'].hist(bins=20,alpha = 0.5)
stars2['Discounted_Price'].hist(bins=20,color = sns.desaturate('green',0.8),alpha = 0.5)
plt.show()
```

```python
df3 = stars1
```

```python
df3 = df3.append(stars2)
```

```python
df3
```

```python
differences = np.zeros(5000)
for i in np.arange(5000):
    age = df3[['Stars', 'Discounted_Price']]
    shuffled = age.sample(195,replace = False)
    shuffled_ages = shuffled['Discounted_Price']
    original_and_shuffled = age.assign(shuffled_ages=shuffled_ages.values )
    shuffled1 = original_and_shuffled.loc[(original_and_shuffled['Stars'] >= 4.5) & (original_and_shuffled['Stars'] <= 5)]
    shuffled2 = original_and_shuffled.loc[(original_and_shuffled['Stars'] >= 4.3) & (original_and_shuffled['Stars'] <= 4.4)]
    difference = shuffled2['shuffled_ages'].mean() - shuffled1['shuffled_ages'].mean()
    differences[i] = difference
```

```python
differences
```

```python
differences_df = pd.DataFrame(differences)
differences_df
```

```python
differences = pd.DataFrame(differences)
differences.hist(bins=20,color = sns.desaturate('green',0.8),alpha = 0.5)
plt.scatter(observed_difference,0,color = 'red')
```

Fail to reject the null hypothesis
